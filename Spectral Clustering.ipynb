```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.metrics.cluster import adjusted_rand_score, normalized_mutual_info_score, homogeneity_completeness_v_measure
SC = SpectralClustering(n_clusters=2, assign_labels='discretize', random_state=0).fit(X)
y_SC = SC.fit_predict(X)
sil_score = silhouette_score(X, y_SC)
print("Silhouette Score:", sil_score)
ch_score = calinski_harabasz_score(X, y_SC) 
print("Calinski-Harabasz Index:", ch_score)
db_score = davies_bouldin_score(X, y_SC) 
print("Davies-Bouldin Index:", db_score)   
```
