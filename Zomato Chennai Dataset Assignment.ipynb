```python
dataset['Dining Rating'].fillna(0, inplace=True)
dataset['Dining Rating Count'].fillna(0, inplace=True)
dataset['Delivery Rating'].fillna(0, inplace=True)
dataset['Delivery Rating Count'].fillna(0, inplace=True)
```

```python
# Dining Rating Count
dataset['Dining Rating Count'] = (
    dataset['Dining Rating Count']
    .replace({
        'Does not offer Dining': '0',
        'Not enough Dining Reviews': '0',
        'None': '0'
    })
    .fillna('0')
    .astype(str)
    .str.replace(',', '', regex=True)
    .astype('Int64')
)

# Delivery Rating Count
dataset['Delivery Rating Count'] = (
    dataset['Delivery Rating Count']
    .replace({
        'Does not offer Delivery': '0',
        'Not enough Delivery Reviews': '0',
        'None': '0'
    })
    .fillna('0')
    .astype(str)
    .str.replace(',', '', regex=True)
    .astype('Int64')
)

# Dining Rating
dataset['Dining Rating'] = (
    dataset['Dining Rating']
    .replace({
        'Does not offer Dining': '0',
        'Not enough Dining Reviews': '0',
        'None': '0'
    })
    .fillna(0)
    .astype(float)
)

# Delivery Rating
dataset['Delivery Rating'] = (
    dataset['Delivery Rating']
    .replace({
        'Does not offer Delivery': '0',
        'Not enough Delivery Reviews': '0',
        'None': '0'
    })
    .fillna(0)
    .astype(float)
)
```

```python
dataset
```

```python
# 1. Trim spaces and standardize text case
for col in dataset.select_dtypes(include=['object']).columns:
    dataset[col] = dataset[col].str.strip()
```

```python
# 2. Merge similar location names
location_replacements = {
    'Velachery East': 'Velachery',
    'West Mambalam': 'Mambalam',
    'Adyar Beach': 'Adyar',
    'Perungudi Toll Gate': 'Perungudi',
    'Anna Nagar West': 'Anna Nagar',
    'Anna Nagar East': 'Anna Nagar',
    'Ekkaduthangal': 'Guindy',
    'Kotturpuram': 'Adyar',
    'Pallikaranai Medavakkam': 'Pallikaranai',
    'Sholinganallur': 'OMR',
    'Thoraipakkam': 'OMR',
    'Navallur': 'OMR'
}
dataset['Location'] = dataset['Location'].replace(location_replacements, regex=False)
```

```python
dataset['Cuisine'] = (
    dataset['Cuisine']
    .str.replace('Sea food', 'Seafood', regex=False)
    .str.replace('Icecreams', 'Ice Creams', regex=False)
    .str.replace('Biryani ', 'Biryani', regex=False)
)# 4. Fill missing prices with median for each location
dataset['Price for 2'] = dataset.groupby('Location')['Price for 2'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# 5. Remove duplicates (Name + Address)
dataset.drop_duplicates(subset=['Name of Restaurant', 'Address'], inplace=True)
```

```python
# 6. Reorder columns
dataset = dataset[[
    'Name of Restaurant', 'Address', 'Location', 'Cuisine', 'Top Dishes',
    'Price for 2', 'Dining Rating', 'Dining Rating Count',
    'Delivery Rating', 'Delivery Rating Count', 'Features', 'Zomato URL'
]]
```

```python
# 7. Save cleaned dataset
dataset.to_csv("Zomato_Chennai_Cleaned.csv", index=False)

print("✅ Zomato Chennai dataset cleaned and saved as Zomato_Chennai_Cleaned.csv")
```

```python
import plotly.express as px

# Count restaurants by Cuisine
cuisine_count = dataset['Cuisine'].value_counts().reset_index()
cuisine_count.columns = ['Cuisine', 'No of Restaurants']

# Create histogram (actually a bar chart works better for counts)
fig = px.bar(
    cuisine_count.head(20),  # top 20 cuisines
    x='Cuisine',
    y='No of Restaurants',
    title='Number of Restaurants by Cuisine (Top 20)',
    text='No of Restaurants',
    color='No of Restaurants',
    color_continuous_scale='Blues'
)

fig.update_layout(xaxis_title='Cuisine', yaxis_title='Number of Restaurants')
fig.show()
```

```python
#Chennai’s location-wise restaurant distribution.

import plotly.express as px

#count resturants by location
location_count = dataset['Location'].value_counts().reset_index()
location_count.columns = ['Location', 'No of Restaurants']

# Plot pie chart (Top 20 locations for clarity)
fig = px.pie(
    location_count.head(20),
    names='Location',
    values='No of Restaurants',
    title='Chennai Location Wise Distribution of Restaurants (Top 20)',
    color_discrete_sequence=px.colors.sequential.Teal
)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()

```

```python
# Step 1: Pick top 12 locations by restaurant count
top_locations = dataset['Location'].value_counts().nlargest(12).index

# Step 2: Keep only restaurants in those top locations
df_top = dataset[dataset['Location'].isin(top_locations)]

# Step 3: Order locations by median dining rating
order = (
    df_top.groupby('Location')['Dining Rating']
    .median()
    .sort_values(ascending=False)
    .index
    .tolist()  # ✅ Convert to list
)

# Step 4: Plot boxplot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_top,
    x='Dining Rating',
    y='Location',
    order=order,
    palette='viridis'
)
plt.title('Dining Rating Distribution by Location (Top 12)', fontsize=14)
plt.xlabel('Dining Rating')
plt.ylabel('Location')
plt.show()
```

```python
#Price Comparision Location wise¶
# Step 1: Pick top 12 locations by restaurant count
top_locations = dataset['Location'].value_counts().nlargest(12).index

# Step 2: Keep only restaurants in those top locations
df_top = dataset[dataset['Location'].isin(top_locations)]

# Step 3: Order locations by median price
order = (
    df_top.groupby('Location')['Price for 2']
    .median()
    .sort_values(ascending=False)
    .index
    .tolist()
)

# Step 4: Plot boxplot for price comparison
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df_top,
    x='Price for 2',
    y='Location',
    order=order,
    palette='coolwarm'
)
plt.title('Price for 2 Comparison by Location (Top 12)', fontsize=14)
plt.xlabel('Price for 2')
plt.ylabel('Location')
plt.show()
```
