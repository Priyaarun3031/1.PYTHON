```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
dataset = pd.read_csv("kidney_disease.csv")
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
#how to identify the numerical columns
num_cols = dataset.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(num_cols)
```

```python
#Handle numerical columns
num_cols = ['id', 'age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo']
median_imputer = SimpleImputer(strategy='median')
dataset[num_cols] = median_imputer.fit_transform(dataset[num_cols])
```

```python
#how to identify the categorical columns
cat_cols = dataset.select_dtypes(include=['object']).columns.tolist()
print(cat_cols)
```

```python
cat_cols = ['rbc', 'pc', 'pcc', 'ba', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']
median_imputer = SimpleImputer(strategy="most_frequent")
dataset[cat_cols] = median_imputer.fit_transform(dataset[cat_cols])
```

```python
for col in cat_cols:
    print(f"{col}: {dataset[col].unique()}")


```

```python
ord_cols = ['appet']
mode_imputer = SimpleImputer(strategy="most_frequent")
dataset[ord_cols] = mode_imputer.fit_transform(dataset[ord_cols])
```

```python
print(dataset.isnull().sum())
```

```python
dataset['pcv'] = pd.to_numeric(dataset['pcv'], errors='coerce')
dataset['wc'] = pd.to_numeric(dataset['wc'], errors='coerce')
dataset['rc'] = pd.to_numeric(dataset['rc'], errors='coerce')


```

```python
print(dataset.dtypes)
```

```python
for col in dataset.select_dtypes(include='object').columns:
    print(f"{col}: {dataset[col].unique()}")
```

```python
dataset['dm'] = dataset['dm'].str.strip().str.lower().replace({'\tno': 'no', '\tyes': 'yes'})
```

```python
for col in dataset.select_dtypes(include='object').columns:
    print(f"{col}: {dataset[col].unique()}")
```

```python
dataset['dm'] = dataset['dm'].str.strip().str.lower().replace({'\tno': 'no', '\tyes': 'yes', 'ckd\t': 'ckd'})
```

```python
print(dataset.dtypes)
```

```python
for col in dataset.select_dtypes(include='object').columns:
    dataset[col] = dataset[col].str.strip()
```

```python
for col in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']:
    print(f"{col}: {dataset[col].unique()}")
```

```python
for col in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']:
    dataset[col] = dataset[col].str.lower()
```

```python
dataset['classification'] = dataset['classification'].replace({'ckd\t': 'ckd'})
```

```python
for col in ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']:
    print(f"{col}: {dataset[col].unique()}")
```
